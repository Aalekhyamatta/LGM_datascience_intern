```python
import matplotlib.pyplot as plt
grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_image)
```

```python
inverted_image=255-grey_image
plt.imshow(inverted_image)
```

# smoothing the image

```python
img_smoothing = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
img_smoothing
```

```python
plt.imshow(img_smoothing)
```

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256.0)
```

# 6. Inverting the Image to get the Pencil Sketch

```python
pencil_sketch = dodgeV2(grey_image, img_smoothing)
plt.imshow(pencil_sketch)
```
